```python
clean_titanic = pd.read_csv('data/cleaned_data.csv')
```

```python
o_h_sex =[]
for i in range(len(clean_titanic['Sex'])):
    if clean_titanic['Sex'].iloc[i] == 'male':
        o_h_sex.append(0)
    else:
        o_h_sex.append(1)

clean_titanic['sex_one_hot'] = o_h_sex

# Using dataframe
# X_train.loc[X_train['Sex'] == 'male', 'Sex'] = 0
# X_train.loc[X_train['Sex'] == 'female', 'Sex'] = 1
```

```python
clean_titanic
```

```python
port_dummy = pd.get_dummies(clean_titanic.Port_name, prefix='port')
class_dummy = pd.get_dummies(clean_titanic.Pclass, prefix='class')
titanic_port = pd.concat([clean_titanic, port_dummy], axis=1)
titanic_dummy = pd.concat([titanic_port, class_dummy], axis=1)
titanic_dummy
```

```python
titanic_data = titanic_dummy.drop(['Ticket', 'Sex', 'Fare', 'Port_name', 'Pclass', 'Unnamed: 0'], axis=1)
```

```python
titanic_data
```

```python
from sklearn.model_selection import train_test_split
y = titanic_data.Survived
X = titanic_data.drop(['Survived'], axis=1)

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
DT_model = DecisionTreeRegressor(max_leaf_nodes=4)
```

```python
DT_model.fit(train_X, train_y)
from sklearn.metrics import mean_absolute_error

val_predictions = DT_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
val_mae
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]
# Write loop to find the ideal tree size from candidate_max_leaf_nodes
error = 100000
for leaf_dept in candidate_max_leaf_nodes:
    my_mae = get_mae(leaf_dept, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(leaf_dept, my_mae))
    if error > my_mae:
        error = my_mae
        opt_leaf_struct = leaf_dept
    
# Store the best value of max_leaf_nodes (it will be either 5, 25, 50, 100, 250 or 500)
best_tree_size = opt_leaf_struct
print(best_tree_size)
```
